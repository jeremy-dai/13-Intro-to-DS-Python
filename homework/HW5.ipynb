### Exercise 2: Undersampling
Loading the Sinterklaas image should not be a problem. However, some images are too big to be processed easily on your computer. When an image is big, the resolution can be adjusted by undersampling your image. Undersampling means that instead of working with the full image, we only use every N'th column and row as our image. This can preserve the original image, while lowering the memory required to store the image. In this excercise we will undersample the 'Sinterklaas.jpg' image by a factor of N (N'th pixel!) in both the x and y directions after having loaded the entire image. Make sure that you start at index `0` and include all N'th pixels (including the last pixel if it it divisible by N).

```python
#undersample
# in: numpy array with 3 dimensions: image
#     integer containing the undersampling factor: N
# out: undersampled numpy array with 3 dimensions

def undersample(image, N):
    #Complete your function here
    return image[::N,::N,:]
```

```python
#summary_pixels
# in: two dimensional numpy array: greyscale_image
# out: tuple containing the mean and standard deviation of the pixel distribution over 125
def summary_pixels(grayscale_image):
    up = grayscale_image.flat[grayscale_image.flat > 125]
    return (up.mean(),up.std())
```

### Excercise 5: Tiling

In this excercise we will tile a bigger array with the greyscale image of Sinterklaas. Tiling is the process of filling an array with the same scalars or subarray entirely by repeating the values you put in. An example of an array tiled with the scalar 3 is given below:

$$\pmatrix{3&3&3&3\\3&3&3&3\\3&3&3&3}$$

In our case we will tile the image with the greyscaled Sinterklaas image and a transposed version of this image in a checkers board pattern. If we let $S$ be the original greyscale image and $T$ the transposed version, this should look something like this:

$$\pmatrix{S&T&S&T\\T&S&T&S\\S&T&S&T}$$

You are going to write a function that can output a tiled numpy array (or, greyscale image) given the number of tiles in rows and columns. Make sure that your code can also work with different (square) images.

```python
#tiling
# in: two-dimensional numpy array: greyscale_image
#     integer with number of columns of tiles: x_tiles
#     integer with number of rows of tiles: y_tiles
# out: two dimensional numpy array with x_tiles by y_tiles tiles with transpose checkered pattern

def tiling(greyscale_image, x_tiles, y_tiles):
    _col = [greyscale_image,greyscale_image.T]
    col1 = _col[0]
    col2 = _col[1]
    for i in range(x_tiles-1):
        col1 = np.vstack([col1,_col[(i+1)%2]])
        col2 = np.vstack([col2,_col[(i)%2]])
    _row = [col1,col2]
    final = _row[0]
    for i in range(y_tiles-1):
        final = np.hstack([final,_row[(i+1)%2]])
    
    return final
```
